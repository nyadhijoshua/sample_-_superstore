```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate some sample correlation data for demonstration purposes
data = pd.DataFrame(np.random.rand(5, 5))

# Calculate correlation matrix
corr_matrix = data.corr()

# Set the figure size
plt.rcParams["figure.figsize"] = (10, 7)

# Create the heatmap using imshow
heatmap = plt.imshow(corr_matrix, cmap='coolwarm')

# Add annotations with correlation values
for i in range(len(corr_matrix)):
    for j in range(len(corr_matrix)):
        text = f"{corr_matrix.iloc[i, j]:.2f}"
        plt.text(j, i, text, ha='center', va='center', color='w', fontsize=12)

# Show colorbar
plt.colorbar(heatmap)

# Show the plot
plt.show()
```

```python

# Create a correlation matrix
corr = np.corrcoef(data.values)

# Create a heatmap
plt.imshow(corr, cmap='Blues')
plt.title('Correlation Heatmap')
plt.colorbar()
plt.show()
```
